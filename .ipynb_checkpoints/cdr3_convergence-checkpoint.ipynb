```python
os.chdir('../..')
```

```python
config = pd.read_csv("data/10X_bcr_vdj_example/sc5p_v2_hs_B_1k_b/filtered_contig_annotations.csv")
config
```

```python
cdr3_h = config[config['chain'] == "IGH"]
cdr3_h
```

```python
cdr3_h.columns
```

```python
v_gene_counts = cdr3_h.groupby('v_gene').count()['barcode'].reset_index().rename(columns={"barcode": "counts"})
v_gene_counts = v_gene_counts.sort_values(by=['counts'], ascending=False)

plt.figure(figsize=(4, 8))
sns.barplot(data=v_gene_counts,
           y="v_gene", x="counts");
```
