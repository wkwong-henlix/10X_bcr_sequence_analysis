## EDA

```python
antihen_umi = antigen_score[antigen_score['antigen'] == 'Anti-Hen_Egg_Lysozyme'][['barcode', 'antigen_umi']]
antihen_umi = antihen_umi.rename(columns={'antigen_umi': 'anti-hen_umi'}).reset_index(drop=True)

gp120_umi = antigen_score[antigen_score['antigen'] == 'gp120'][['barcode', 'antigen_umi']]
gp120_umi = gp120_umi.rename(columns={'antigen_umi': 'gp120_umi'})[['gp120_umi']].reset_index(drop=True)

control_umi = antigen_score[['barcode', 'control_umi']].drop_duplicates()[['control_umi']].reset_index(drop=True)

umi_df = pd.concat([antihen_umi, gp120_umi, control_umi], axis=1).set_index("barcode").join(airr.set_index("cell_id"), how="inner")
umi_df
```

```python
sns.histplot(umi_df['sequence_aa'].apply(len))
plt.title("Length of Ig Chain Sequence");
```

```python
sns.histplot(data=umi_df['anti-hen_umi'], kde=True)

plt.title("Predictor Distribution");
```

```python
sns.histplot(data=np.log(umi_df['anti-hen_umi']+1), kde=True)

plt.title("Predictor Distribution, Log Transformed");
```

```python
mean = np.mean(umi_df['anti-hen_umi'])
std = np.std(umi_df['anti-hen_umi'])

sns.histplot(data=(umi_df['anti-hen_umi']-mean)/std, kde=True)

plt.title("Predictor Distribution, Normalized");
```

```python
# Create a df for the train test split
prediction_df = umi_df.copy()

# Convert response variable into log
prediction_df['anti-hen_umi_log'] = np.log(umi_df['anti-hen_umi']+1)
prediction_df = prediction_df[['sequence_aa', 'anti-hen_umi_log']].reset_index()

# Randomly sample for balanced umi dataset
prediction_df_balanced = pd.DataFrame()

for umi_log in np.arange(1, np.max(prediction_df['anti-hen_umi_log'])):
    selected_umi = prediction_df[prediction_df['anti-hen_umi_log'].between(umi_log-1, umi_log, inclusive = True)]
    sample = selected_umi.sample(n=200, replace=True)
    prediction_df_balanced = pd.concat([prediction_df_balanced, sample], axis=0)

prediction_df_balanced 
```
